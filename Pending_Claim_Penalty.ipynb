```python
import time
from time import sleep
from selenium import webdriver
from getpass import getpass
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import Select
from selenium.common.exceptions import TimeoutException
```

```python
files = 0
district_menu = Select(WebDriverWait(driver,50).until(EC.visibility_of_element_located((By.XPATH, '//*[@id="app"]/div/div[1]/div[2]/div/div[2]/div[2]/div/div/div[1]/div[2]/div/div/div[1]/div/select'))))
for index in range(1, len(district_menu.options)):
    district_menu.select_by_index(index)
    district_options = district_menu.options
    selected_district = district_options[index].text
    WebDriverWait(driver, 500).until(EC.invisibility_of_element_located((By.XPATH, "//div[@class='loader__loader___tH-dK']")))
    
    crop_menu = Select(WebDriverWait(driver,20).until(EC.visibility_of_element_located((By.XPATH, '//*[@id="app"]/div/div[1]/div[2]/div/div[2]/div[2]/div/div/div[1]/div[2]/div/div/div[2]/div/select'))))
    for index in range(1, len(crop_menu.options)):
        crop_menu.select_by_index(index)
        crop_options = crop_menu.options
        selected_crop = crop_options[index].text
        WebDriverWait(driver, 500).until(EC.invisibility_of_element_located((By.XPATH, "//div[@class='loader__loader___tH-dK']")))

        element = driver.find_element(By.XPATH, '//*[@id="app"]/div/div[1]/div[2]/div/div[2]/div[2]/div/div/div[2]/div[1]/span/strong')
        value_text = element.text
        value = int(value_text)

        if value == 0:
            print(f"------ No data found for district- {selected_district}, crop- {selected_crop}")
            
        elif value < 150000:
            WebDriverWait(driver, 2).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="app"]/div/div[1]/div[2]/div/div[2]/div[2]/div/div/div[3]/div[2]/div/button'))).click()
            WebDriverWait(driver, 500).until(EC.invisibility_of_element_located((By.XPATH, "//div[@class='loader__loader___tH-dK']")))
            print(f"Data downloaded for district- {selected_district}, crop- {selected_crop}, apps- {value}")
            files = files + 1

        elif value >= 150000:
            iu_menu = Select(WebDriverWait(driver,20).until(EC.visibility_of_element_located((By.XPATH, '//*[@id="app"]/div/div[1]/div[2]/div/div[2]/div[2]/div/div/div[1]/div[2]/div/div/div[3]/div/select'))))
            options = iu_menu.options
            print(f"...Downloading IU-wise for district- {selected_district}, crop- {selected_crop}, apps- {value}")
            
            for index in range(1, len(iu_menu.options)):                
                iu_menu.select_by_index(index)
                WebDriverWait(driver, 500).until(EC.invisibility_of_element_located((By.XPATH, "//div[@class='loader__loader___tH-dK']")))

                iu_apps = driver.find_element(By.XPATH, '//*[@id="app"]/div/div[1]/div[2]/div/div[2]/div[2]/div/div/div[2]/div[1]/span/strong')
                iu_value = int(iu_apps.text)

                if iu_value == 0:
                    continue
                else:
                    WebDriverWait(driver, 2).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="app"]/div/div[1]/div[2]/div/div[2]/div[2]/div/div/div[3]/div[2]/div/button'))).click()
                    WebDriverWait(driver, 500).until(EC.invisibility_of_element_located((By.XPATH, "//div[@class='loader__loader___tH-dK']")))
                    files = files + 1
                
                print(f"Data downloaded for district- {selected_district}, crop- {selected_crop}, apps- {value}")
                
            WebDriverWait(driver, 30).until(EC.element_to_be_clickable((By.XPATH, "//*[@id='app']/div/div[1]/div[2]/div/div[1]/div/div/div/ul/li[2]/a"))).click()
            WebDriverWait(driver, 30).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="app"]/div/div[1]/div[2]/div/div[2]/div[2]/div/div/div/div[3]/div[5]/div/a'))).click()
            district_menu = Select(WebDriverWait(driver,50).until(EC.visibility_of_element_located((By.XPATH, '//*[@id="app"]/div/div[1]/div[2]/div/div[2]/div[2]/div/div/div[1]/div[2]/div/div/div[1]/div/select'))))
            district_menu.select_by_visible_text(selected_district)
            crop_menu = Select(WebDriverWait(driver,20).until(EC.visibility_of_element_located((By.XPATH, '//*[@id="app"]/div/div[1]/div[2]/div/div[2]/div[2]/div/div/div[1]/div[2]/div/div/div[2]/div/select'))))
            crop_menu.select_by_visible_text(selected_crop)
            WebDriverWait(driver, 500).until(EC.invisibility_of_element_located((By.XPATH, "//div[@class='loader__loader___tH-dK']")))
            
        else:
            print(f"ERROR for district- {selected_district}, crop- {selected_crop}")
```
